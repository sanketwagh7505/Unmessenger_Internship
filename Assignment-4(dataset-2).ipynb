### Data Visualization

#### 1. The Pie Chart Showing the Gender Ratio in the Data

```python
dataset['Sex'] = dataset['Sex'].replace({'M':"Male", 'F': "Female"})
gender_counts = dataset['Sex'].value_counts()
```

    - In the above cell we are replacing M & F with Male & Female to get the clear understanding of data. 
    - Then Counting the Value for each gender and storing it in gender_counts. 

```python
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%')
plt.title('Sex Ratio')
plt.show()
```

    - From the Above Pie chart we can Determine the ratio of Male & Female in the Company. 
    - The autopct keyword in the pie() function is used to display the percentage value on each wedge of the pie chart

#### 2. The Count Plot showing Employee satisfaction levels in the company. 

```python
sns.countplot(x='EmpSatisfaction', data=dataset, palette='viridis')
plt.xlabel('Employee Satisfaction Score')
plt.ylabel('Count')
plt.title('Employee Satisfaction Count')
plt.show()
```

    - The above Count plot displays the Employee satisfaction level in terms of rating.
    - The palette is used for color maping in which viridis is the colormaping has been chosen. 

#### 3. Bar Plot showcasing the average salary paid to employees by their respective job Positions.

```python
avg_salary_by_position = dataset.groupby('Position')['Salary'].mean().reset_index()

sns.barplot(x='Salary', y='Position', data=avg_salary_by_position)
plt.title('Average Salary by Position')
plt.show()
```

    - The above barplot showcases the average salary for different Job Positions
    - In avg_salary_by_position 
            - we grouped the data by Position using groupby
            - then, we calculated the mean of Salary 
            - and finally we reseted the indexes of it after grouping it by Position.

#### 4. Violin Plot showing the distribution of salary based on performace Score.

```python
sns.violinplot(x='PerformanceScore', y='Salary', data=dataset)
plt.title('Salary Distribution by PerformanceScore')
plt.show()
```

    - The Violin Plot above shows the corelation between Performance Score of the employee & the salary of the employee.

#### 5. Line Plot Showcasing Salary of Employees over time

```python
dataset.sort_values('DateofHire', inplace=True)
date_range = (dataset['DateofHire'].dt.year >= 2014) & (dataset['DateofHire'].dt.year <= 2016)
date_filtered = dataset.loc[date_range]
```

    - First we sort the date in Date of Hire Column 
    - Then we create the start year & end year by using and formula in it.
    - Then to locate the given start year & end year in date_range in Date of Hire column we use loc & give date_range variable in it 
    - Then store it in data_filtered variable

```python
plt.figure(figsize=(15, 4))
plt.plot(date_filtered['DateofHire'], date_filtered['Salary'])
plt.xlabel('Date of Hire')
plt.ylabel('Salary')
plt.title('Salary over Time')
plt.show()
```

    - In the above Line Plot we are visualizing the salary of employees Over Time

#### 6. Stacked Bar Plot showing Employee Status by Department

```python
df_grouped = dataset.groupby(['Department', 'EmploymentStatus']).size().unstack()
df_grouped.plot(kind='bar', stacked=True)
plt.xlabel('Department')
plt.ylabel('Count')
plt.title('Employee Status by Department')
plt.show()
```

    - The Above stack bar plot displays the status of all the employees in different departments.
    - The employees are sorted into groups based on their department & Employees status by groupby function.
    - Each bar represents a unique department and its corresponding employee's status is represented by color.
    - the size counts how many employees are in each group.
    - the unstack is used to rearrange the data by creating new table where each department is a row & each employment status is column by doing this the data becomes more readable.
    - the kind shows what type of graph we are using.
    - the stacked is used to make the representation in stacked bar chart.


#### 7. Histogram Plot showing the distribution of Age Distribution

```python

pivot_table = dataset.pivot_table(values='EmpSatisfaction', index='PerformanceScore', aggfunc='mean')

sns.heatmap(pivot_table, cmap='coolwarm')
plt.title('Heatmap of Employee Satisfaction and Performance Score')
plt.show()
```

    - The above heat map visualizes the Performance Score & Employee Satisfaction along with the scale to represent it.
    - First we create a pivot_table to pass in the heatmap.
    - the aggfunc is calculating the average or the mean employee satisfaction for each performance Score.
    - the cmap is color mapping, which will be used to represent different values of the data.
